```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r"C:\Users\DELL\Desktop\project data\cleaned_data.csv")

data.head()
```

```python
columns_to_analyse=['Survived','Pclass','Age','SibSp','Parch','Ticket','Fare']
```

```python
mean_values=data[columns_to_analyse].mean()
median_values=data[columns_to_analyse].median()
mode_values=data[columns_to_analyse].mode().iloc[0]
std_dev_values=data[columns_to_analyse].std()

print("Mean:\n",mean_values)
print("Median:\n",median_values)
print("Mode:\n",mode_values)
print("Standard Deviation:\n",std_dev_values)
```
