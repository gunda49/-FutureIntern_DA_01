```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\DELL\Desktop\project data\Iris.csv")

df.head()
```

```python
df.info()
```

```python
#Histogram for SepalLengthCm
plt.hist(df['SepalLengthCm'], bins=20)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Lenght')
plt.show()
```

```python
#Histogram for SepalWidthCm
plt.hist(df['SepalWidthCm'], bins=20)
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Width')
plt.show()
```

```python
#Histogram for PetalLengthCm
plt.hist(df['PetalLengthCm'], bins=20)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Length')
plt.show()
```

```python
#Histogram for PetalWidthCm
plt.hist(df['PetalWidthCm'], bins=20)
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Width')
plt.show()
```
